# Data Cleaning

```python
def drop_cols_w_missing_values(df: pd.DataFrame, threshold: float) -> pd.DataFrame:
    """Drop columns with missing values above a certain threshold."""
    missing_fraction = df.isnull().mean()
    cols_to_drop = missing_fraction[missing_fraction > threshold].index
    # Print dropped columns for reference
    print(f"Dropping columns with more than {threshold*100}% missing values")
    print(cols_to_drop.tolist())
    return df.drop(columns=cols_to_drop)
```

```python
DROP_THRESHOLD = 0.9
df_cleaned = drop_cols_w_missing_values(train_data, DROP_THRESHOLD)
```

```python
df_cleaned.shape
```

```python
def drop_cols_w_one_value(df: pd.DataFrame) -> pd.DataFrame:
    """Drop columns that contain only a single unique value."""
    cols_to_drop = [col for col in df.columns if df[col].nunique() <= 1]
    # Print dropped columns for reference
    print("Dropping columns with a single unique value")
    print(cols_to_drop)
    return df.drop(columns=cols_to_drop)
```

```python
df_cleaned = drop_cols_w_one_value(df_cleaned)
```

```python
# Drop transacrionID col
df_cleaned = df_cleaned.drop(columns=['TransactionID'])

# Sort data by 'TransactionDT' and split into train and validation sets
df_sorted = df_cleaned.sort_values(by='TransactionDT').reset_index(drop=True)

split_index = int(len(df_sorted) * 0.8)  # Split 80-20
train_set = df_sorted.iloc[:split_index]
validation_set = df_sorted.iloc[split_index:]
```

# Baseline Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import precision_recall_curve, auc
```

```python
# separating target from features
train_X = train_set.drop(columns=[TARGET_COL])
train_Y = train_set[TARGET_COL]

val_X = validation_set.drop(columns=[TARGET_COL])
val_Y = validation_set[TARGET_COL]
```

```python
'''
Preprocessing to include scaling for numeric features,
and one-hot encoding for categorical features.
Also handle NaN values using median and mode imputation.
'''
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, NUM_COLS),
        ('cat', categorical_transformer, CAT_COLS)
    ])

baseline_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])
baseline_model.fit(train_X, train_Y)
```

```python
from sklearn.metrics import (
    average_precision_score,
    precision_recall_curve,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

# 1) Predict probabilities on validation set
val_proba = baseline_model.predict_proba(val_X)[:, 1]

# 2) PR-AUC (Average Precision) - primary metric for imbalance
pr_auc = average_precision_score(val_Y, val_proba)
print(f"PR-AUC (Average Precision): {pr_auc:.4f}")

# 3) Choose a threshold that maximizes F1 on validation
precision, recall, thresholds = precision_recall_curve(val_Y, val_proba)
f1_scores = 2 * precision * recall / (precision + recall + 1e-12)

# thresholds has length = len(precision)-1, so align indices
best_idx = np.argmax(f1_scores[:-1])
threshold = thresholds[best_idx]
print(f"Chosen threshold (max F1): {threshold:.4f}")

# 4) Convert probabilities to predicted labels
val_pred = (val_proba >= threshold).astype(int)

# 5) Compute precision / recall / F1
prec = precision_score(val_Y, val_pred, zero_division=0)
rec = recall_score(val_Y, val_pred)
f1 = f1_score(val_Y, val_pred)

print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-score:  {f1:.4f}")

# 6) Confusion matrix
cm = confusion_matrix(val_Y, val_pred)
print("Confusion Matrix [[TN FP],[FN TP]]:")
print(cm)

# 7) Full classification report
print("\nClassification report:")
print(classification_report(val_Y, val_pred, digits=4, zero_division=0))
```
